# Audio classifier: Data Preprocessing

```python
# !pip install SoundFile
import csv
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import pandas as pd
from pathlib import Path
from scipy import signal
```

```python
meta = {"filename": [], "fold": [], "target": [], "category": [], "esc10": [], "src_file": [], "take": []}
with open('CSV.csv', newline='') as csv_file:
    reader = csv.reader(csv_file, delimiter=',')
    for row in reader:
        meta["filename"].append(row[0])
        meta["fold"].append(row[1])
        meta["target"].append(row[2])
        meta["category"].append(row[3])
        meta["esc10"].append(row[4])
        meta["src_file"].append(row[5])
        meta["take"].append(row[6])

# convert dict to dataframe
df_meta = pd.DataFrame(meta)
df_meta
```

```python
selected_classnames = [ classname.lower() for classname in [
'Breathing',
'Coughing',
'Footsteps',
'Laughing',
'Pouring water',
'Sneezing',
'Snoring',
'Toilet_flush',
'Vacuum_cleaner',
'Washing_machine',
]]
```

```python
selected_data = (df_meta.loc[df_meta['category'].isin(['breathing', 'coughing', 'footsteps', 'laughing', 'pouring water', 'sneezing', 'snoring', 'toilet_flush', 'vacuum_cleaner', 'washing_machine'])])
```
